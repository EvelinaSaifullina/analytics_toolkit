
#### ПОЛНЫЙ НАБОР ШАБЛОНОВ ДЛЯ ГРАФИКОВ В MATPLOTLIB



```python


# ========== 1. ИМПОРТЫ И НАСТРОЙКИ ==========
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Стиль и настройки (можно менять по вкусу)
plt.style.use('seaborn-v0_8-whitegrid')   # красивый стиль
plt.rcParams['figure.figsize'] = (12, 6)  # размер по умолчанию
plt.rcParams['font.size'] = 12

# ========== 2. СОЗДАЁМ ТЕСТОВЫЕ ДАННЫЕ ==========
np.random.seed(42)  # для воспроизводимости

# Даты для временного ряда
dates = pd.date_range('2024-01-01', periods=100)

# Данные для графиков
df = pd.DataFrame({
    'date': dates,
    'revenue': np.random.normal(1000, 200, 100).cumsum() + 5000,  # тренд + шум
    'cost': np.random.normal(800, 150, 100).cumsum() + 4000,
    'advertising': np.random.randint(200, 800, 100),
    'age': np.random.randint(18, 70, 200),
    'category': np.random.choice(['A', 'B', 'C', 'D'], 200, p=[0.3, 0.4, 0.2, 0.1])
})

# Добавим колонку с выручкой по категориям (для столбчатой диаграммы)
category_revenue = df.groupby('category')['revenue'].sum()

# ========== 3. ЛИНЕЙНЫЙ ГРАФИК (ТРЕНД) ==========
plt.figure(figsize=(12, 6))
plt.plot(df['date'], df['revenue'], label='Выручка', color='blue', linewidth=2)
plt.plot(df['date'], df['cost'], label='Себестоимость', color='red', linestyle='--')
plt.title('Динамика выручки и себестоимости', fontsize=14)
plt.xlabel('Дата')
plt.ylabel('Рубли')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.3)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# ========== 4. СТОЛБЧАТАЯ ДИАГРАММА (СРАВНЕНИЕ) ==========
plt.figure(figsize=(10, 6))
plt.bar(category_revenue.index, category_revenue.values, color='skyblue', edgecolor='black')
plt.title('Выручка по категориям')
plt.xlabel('Категория')
plt.ylabel('Выручка')
plt.grid(axis='y', linestyle='--', alpha=0.3)
plt.tight_layout()
plt.show()

# ========== 5. ГИСТОГРАММА (РАСПРЕДЕЛЕНИЕ) ==========
plt.figure(figsize=(10, 6))
plt.hist(df['age'], bins=20, color='purple', alpha=0.7, edgecolor='black')
plt.title('Распределение возраста клиентов')
plt.xlabel('Возраст')
plt.ylabel('Количество')
plt.grid(axis='y', linestyle='--', alpha=0.3)
plt.xlim(-60, 120) # ограничение оси Х
plt.legend()
plt.tight_layout()
plt.show()

# ========== 6. ТОЧЕЧНЫЙ ГРАФИК (SCATTER) ==========
plt.figure(figsize=(10, 6))
plt.scatter(df['advertising'], df['revenue'], alpha=0.6, color='orange', edgecolor='black')
plt.title('Зависимость выручки от рекламных расходов')
plt.xlabel('Рекламные расходы')
plt.ylabel('Выручка')
plt.grid(True, linestyle='--', alpha=0.3)
plt.tight_layout()
plt.show()

# ========== 7. ЯЩИК С УСАМИ (BOXPLOT) ==========
# Соберём данные по категориям для boxplot
data_by_cat = [df[df['category'] == cat]['revenue'].values for cat in category_revenue.index]
plt.figure(figsize=(10, 6))
plt.boxplot(data_by_cat, labels=category_revenue.index)
plt.title('Распределение выручки по категориям')
plt.xlabel('Категория')
plt.ylabel('Выручка')
plt.grid(axis='y', linestyle='--', alpha=0.3)
plt.tight_layout()
plt.show()

# ========== 8. КРУГОВАЯ ДИАГРАММА (PIE) ==========
plt.figure(figsize=(8, 8))
plt.pie(category_revenue.values, labels=category_revenue.index, autopct='%1.1f%%', startangle=90)
plt.title('Доля выручки по категориям')
plt.axis('equal')
plt.tight_layout()
plt.show()

# ========== 9. НЕСКОЛЬКО ГРАФИКОВ В ОДНОМ ОКНЕ (SUBPLOTS) ==========
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# 1. Линейный
axes[0, 0].plot(df['date'], df['revenue'], color='blue')
axes[0, 0].set_title('Выручка по дням')
axes[0, 0].grid(True)

# 2. Столбчатая
axes[0, 1].bar(category_revenue.index, category_revenue.values, color='green')
axes[0, 1].set_title('Выручка по категориям')
axes[0, 1].grid(True, axis='y')

# 3. Гистограмма
axes[1, 0].hist(df['age'], bins=15, color='purple', alpha=0.7)
axes[1, 0].set_title('Распределение возраста')
axes[1, 0].grid(True, axis='y')

# 4. Scatter
axes[1, 1].scatter(df['advertising'], df['revenue'], alpha=0.5)
axes[1, 1].set_title('Реклама vs Выручка')
axes[1, 1].grid(True)

plt.tight_layout()
plt.show()

# ========== 10. КАК СОХРАНИТЬ ГРАФИК В ФАЙЛ ==========
# Раскомментируй, если нужно сохранить последний график
# plt.savefig('my_plot.png', dpi=300, bbox_inches='tight')

print("✅ Все графики построены! Ты можешь копировать нужный блок в свои проекты.")
```
